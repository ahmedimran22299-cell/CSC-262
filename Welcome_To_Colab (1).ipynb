```
# --- Graph Definition from Image ---
graph = {
    '5' : ['3', '7'],
    '3' : ['2', '4'],
    '7' : ['8'],
    '2' : [],
    '4' : ['8'],
    '8' : []
}

# --- Initialization ---
visited = [] # List for visited nodes.
queue = []   # Initialize a queue

# --- BFS Function Logic ---
def bfs(visited, graph, node): # Function for BFS
    visited.append(node)
    queue.append(node)

    while queue: # Creating loop to visit each node
        m = queue.pop(0)
        print(m, end=" ")

        for neighbour in graph[m]:
            if neighbour not in visited:
                visited.append(neighbour)
                queue.append(neighbour)

# --- Driver Code ---
print("Following is the Breadth-First Search:")
bfs(visited, graph, '5')
```

```
# Function to print the parent of each node using DFS
def printParents(node, adj, parent):
    # current node is Root, thus, has no parent
    if (parent == 0):
        print(node, "-> Root")
    else:
        print(node, "->", parent)

    # Using DFS to traverse and find parents
    for cur in adj[node]:
        if (cur != parent):
            printParents(cur, adj, node)

# Function to print the children of each node using BFS
def printChildren(Root, adj):
    # Queue for the BFS
    q = []
    # pushing the root
    q.append(Root)
    # visit array to keep track of nodes that have been visited
    vis = [0] * len(adj)

    # BFS
    while (len(q) > 0):
        node = q[0]
        q.pop(0)
        vis[node] = 1
        print(node, "-> ", end="")
        for cur in adj[node]:
            if (vis[cur] == 0):
                print(cur, " ", end="")
                q.append(cur)
        print()

# Function to print the leaf nodes
def printLeafNodes(Root, adj):
    # Leaf nodes have only one connection and are not the root
    for i in range(1, len(adj)):
        if (len(adj[i]) == 1 and i != Root):
            print(i, end=" ")
    print()

# Function to print the degrees of each node
def printDegrees(Root, adj):
    for i in range(1, len(adj)):
        print(i, ": ", end="")
        # Root degree is total connections; others are connections minus parent
        if (i == Root):
            print(len(adj[i]))
        else:
            print(len(adj[i]) - 1)

# --- Driver code ---
N = 7
Root = 1

# Adjacency list to store the tree (N+1 to use 1-based indexing)
adj = [[] for _ in range(N + 1)]

# Creating the tree (Undirected edges)
adj[1].append(2); adj[2].append(1)
adj[1].append(3); adj[3].append(1)
adj[1].append(4); adj[4].append(1)
adj[2].append(5); adj[5].append(2)
adj[2].append(6); adj[6].append(2)
adj[4].append(7); adj[7].append(4)

# Execution
print("The parents of each node are:")
printParents(Root, adj, 0)

print("\nThe children of each node are:")
printChildren(Root, adj)

print("\nThe leaf nodes of the tree are:")
printLeafNodes(Root, adj)

print("\nThe degrees of each node are:")
printDegrees(Root, adj)
```

```
class Node:
    def __init__(self, state, parent, actions, cost):
        self.state = state
        self.parent = parent
        self.actions = actions
        self.total_cost = cost

# 1. Define the weighted graph based on the image
graph = {
    'A': Node('A', None, [('B', 6), ('C', 9), ('E', 1)], 0),
    'B': Node('B', None, [('A', 6), ('D', 3), ('E', 4)], 0),
    'C': Node('C', None, [('A', 9), ('F', 2), ('G', 3)], 0),
    'D': Node('D', None, [('B', 3), ('E', 5), ('F', 7)], 0),
    'E': Node('E', None, [('A', 1), ('B', 4), ('D', 5), ('F', 6)], 0),
    'F': Node('F', None, [('C', 2), ('E', 6), ('D', 7)], 0),
    'G': Node('G', None, [('C', 3)], 0)
}

def bfs_with_cost(graph, start, goal):
    queue = [(start, [start], 0)] # (current_node, path_taken, cumulative_cost)
    visited = set()

    while queue:
        (node, path, cost) = queue.pop(0)

        if node not in visited:
            visited.add(node)

            if node == goal:
                return path, cost

            # Explore neighbors (actions)
            for (neighbour, edge_cost) in graph[node].actions:
                if neighbour not in visited:
                    new_path = path + [neighbour]
                    new_cost = cost + edge_cost
                    queue.append((neighbour, new_path, new_cost))

    return None, 0

# --- Driver Code ---
start_state = 'A'
goal_state = 'G'

path, total_cost = bfs_with_cost(graph, start_state, goal_state)

print(f"BFS Path from {start_state} to {goal_state}: {' -> '.join(path)}")
print(f"Total Cost of this path: {total_cost}")
```

```
# BFS Implementation for the Romania Map (Arad to Bucharest)

# i. Corresponding graph based on the figure
romania_map = {
    'Arad': [('Zerind', 75), ('Sibiu', 140), ('Timisoara', 118)],
    'Zerind': [('Arad', 75), ('Oradea', 71)],
    'Oradea': [('Zerind', 71), ('Sibiu', 151)],
    'Sibiu': [('Arad', 140), ('Oradea', 151), ('Fagaras', 99), ('Rimnicu Vilcea', 80)],
    'Timisoara': [('Arad', 118), ('Lugoj', 111)],
    'Lugoj': [('Timisoara', 111), ('Mehadia', 70)],
    'Mehadia': [('Lugoj', 70), ('Drobeta', 75)],
    'Drobeta': [('Mehadia', 75), ('Craiova', 120)],
    'Craiova': [('Drobeta', 120), ('Rimnicu Vilcea', 146), ('Pitesti', 138)],
    'Rimnicu Vilcea': [('Sibiu', 80), ('Craiova', 146), ('Pitesti', 97)],
    'Fagaras': [('Sibiu', 99), ('Bucharest', 211)],
    'Pitesti': [('Rimnicu Vilcea', 97), ('Craiova', 138), ('Bucharest', 101)],
    'Bucharest': [('Fagaras', 211), ('Pitesti', 101), ('Giurgiu', 90), ('Urziceni', 85)],
    'Giurgiu': [('Bucharest', 90)],
    'Urziceni': [('Bucharest', 85), ('Vaslui', 142), ('Hirsova', 98)],
    'Vaslui': [('Iasi', 92), ('Urziceni', 142)],
    'Iasi': [('Neamt', 87), ('Vaslui', 92)],
    'Neamt': [('Iasi', 87)],
    'Hirsova': [('Urziceni', 98), ('Eforie', 86)],
    'Eforie': [('Hirsova', 86)]
}

# ii. Breadth First Search function to find path and total distance
def bfs_romania(graph, start, goal):
    # Queue stores tuples: (current_node, path_list, total_distance)
    queue = [(start, [start], 0)]
    visited = set()

    while queue:
        (node, path, distance) = queue.pop(0)

        if node not in visited:
            visited.add(node)

            # Check if we reached Bucharest
            if node == goal:
                return path, distance

            # Expand neighbors
            for (neighbor, weight) in graph.get(node, []):
                if neighbor not in visited:
                    new_path = list(path)
                    new_path.append(neighbor)
                    new_distance = distance + weight
                    queue.append((neighbor, new_path, new_distance))

    return None, 0

# iii. Driver Code
start_node = 'Arad'
goal_node = 'Bucharest'

path, total_distance = bfs_romania(romania_map, start_node, goal_node)

print("--- BFS Romania Map Traversal ---")
if path:
    print(f"Shortest Path (by hops): {' -> '.join(path)}")
    print(f"Total Distance: {total_distance} km")
else:
    print("No path found.")
```
